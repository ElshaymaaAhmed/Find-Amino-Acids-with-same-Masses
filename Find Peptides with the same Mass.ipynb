### 3) Proteolytic Digestion with Trypsin 

```python
from pyopenms import *

dig = ProteaseDigestion()
dig.getEnzymeName() # Trypsin

# Read all Fasta File Record
bsa = "".join([l.strip() for l in open("uniprot-organism-yes.fasta").readlines() if l.startswith('>') == False])

# Converet to amino acid
bsa = AASequence.fromString(bsa)

# create all digestion products
result = []
dig.digest(bsa, result)
print(result[5].toString())
print(len(result)) # 1217442 peptides
```

```python
# import module
from prettytable import PrettyTable


# Specify the Column Names while initializing the Table
myTable = PrettyTable(["Mass (g/mol)", "Sequence"])

# Add rows
results = result[0:200] # first 200 peptides as the data is so large and it take too much to load you can change it whenever you want

for s in results:
    myTable.add_row([s.getMonoWeight(), s.toString()])

# print the results with mass for each peptide
print(myTable)
```
